```python
class Person:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def display(self):
        print(f'ì´ë¦„ : {self.name}, ì„±ë³„ : {self.gender}\në‚˜ì´ : {self.age}')
    # ë‚˜ì´ê°€ 20ì„¸ ì´ìƒì´ë©´ ì„±ì¸, ì•„ë˜ë©´ í•™ìƒì´ ì¶œë ¥ë˜ëŠ” í•¨ìˆ˜ ì •ì˜
    def greet(self):
        if self.age >= 20:
            print(f'ì•ˆë…•í•˜ì„¸ìš”, {self.name}! ì„±ì¸ì´ì‹œêµ°ìš”!')
        else:
            print(f'ì•ˆë…•í•˜ì„¸ìš”, {self.name}! í•™ìƒì´ì‹œêµ°ìš”!')
```

```python
# ì…ë ¥ê°’ ë³€ìˆ˜ì— ì €ì¥í•œ ë‹¤ìŒ ê°ì²´ ìƒì„±í•˜ê¸°
age = int(input('ë‚˜ì´ : '))
name = input('ì´ë¦„ : ')
while True:
    gender = input('ì„±ë³„ : ')
    # ìœ íš¨ì„± ê²€ì‚¬ì´ë¯€ë¡œ ì´ ë‘˜ ì¤‘ í•˜ë‚˜ì¸ ê²½ìš°ì—ë§Œ ì…ë ¥í•œ ì„±ë³„ì„ return
    if gender in ['female', 'male']:
        break
    # ì•„ë‹ ê²½ìš°ì—ëŠ” ê²½ê³ ë¬¸ ì¶œë ¥
    else:
        print("ì˜ëª»ëœ ì„±ë³„ì„ ì…ë ¥í•˜ì…¨ìŠµë‹ˆë‹¤. 'male' ë˜ëŠ” 'female'ì„ ì…ë ¥í•˜ì„¸ìš”.")
Person1 = Person(name, gender, age)
# ì •ë³´ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
Person1.display()
# ë‚˜ì´ì— ë”°ë¼ ì¸ì‚¬í•˜ëŠ” í•¨ìˆ˜
Person1.greet()
```

```python
# ì„±ë³„ ì˜¤ë¥˜ ì—†ì´ ì…ë ¥ë°›ëŠ” í•¨ìˆ˜ ì„ ì–¸
def get_gender():
    while True:
        gender = input('ì„±ë³„ : ')
        # ìœ íš¨ì„± ê²€ì‚¬ì´ë¯€ë¡œ ì´ ë‘˜ ì¤‘ í•˜ë‚˜ì¸ ê²½ìš°ì—ë§Œ ì…ë ¥í•œ ì„±ë³„ì„ return
        if gender in ['male', 'female']:
            return gender
        # ì•„ë‹ ê²½ìš°ì—ëŠ” ê²½ê³ ë¬¸ ì¶œë ¥
        else:
            print("ì˜ëª»ëœ ì„±ë³„ì„ ì…ë ¥í•˜ì…¨ìŠµë‹ˆë‹¤. 'male' ë˜ëŠ” 'female'ì„ ì…ë ¥í•˜ì„¸ìš”.")
# ê°ì²´ ë‚´ì—ì„œ ê°’ ì…ë ¥ë°›ê¸°, ê°ì²´ ìƒì„±í•˜ë©´ì„œ ë°”ë¡œ ì…ë ¥ë°›ëŠ” ê²½ìš°
Person2 = Person(name = input('ì´ë¦„ : '), gender = get_gender(), \
                age = int(input('ë‚˜ì´ : ')))

# ì •ë³´ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
Person2.display()
# ë‚˜ì´ì— ë”°ë¼ ì¸ì‚¬í•˜ëŠ” í•¨ìˆ˜
Person2.greet()
```
